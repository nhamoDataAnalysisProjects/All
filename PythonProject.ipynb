```python
# import necessary libraries
import pandas as pd
import seaborn as sns
import numpy as np


import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] =  (12,8)
```

```python
# read the dataset
dataset1 = pd.read_csv(r'/Users/nemo/Downloads/movies.csv')
```

```python
# A look at the first 10 columns of the data
dataset1.head(10)
```

```python
 # find any missing values 

for col in dataset1.columns:
    pct_missing = np.mean(dataset1[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# data types for the columns
dataset1.dtypes
```

```python
# change the data types for some columns

dataset1["budget"] = dataset1["budget"].fillna(0).astype("int64")

dataset1["gross"] = dataset1["gross"].fillna(0).astype("int64")
```

```python
dataset1
```

```python
#Sort by highest revenue
dataset1.sort_values(by=["gross"], inplace = False, ascending = False)
```

```python
# displaying first 10 rows
pd.set_option('display.max_rows', 10)
dataset1
```

```python
#drop any duplicates in out dataset
dataset1['company'].drop_duplicates().sort_values(ascending = False)
```

```python
dataset1
```

```python
# tying to find correlation in dataset with budget vs gross
plt.scatter(x= dataset1['budget'], y = dataset1['gross'])
plt.title('BUDGET VS. GROSS')
plt.xlabel('Gross earnings')
plt.ylabel('Budget for the film')
plt.show()
```

```python
#plot the budget vs gross using seaborn
sns.regplot(x='budget', y = 'gross', data=dataset1, scatter_kws ={'color': 'red'}, line_kws = {'color': 'blue'})
plt.show()
```

```python
#selcting flota type columns to find xorrelation using pearson
numeric_data = dataset1.select_dtypes(include=['float64', 'int64'])
numeric_data.corr(method = 'pearson')
```

```python
# creating a correlation matrix for numeric data type columns
correlation_matrix = numeric_data.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot= True)
plt.title('Correlation matrix for Movie columns')
plt.show()
```

```python
#numerize object coulmns to find correlation in all columns
dataset1_numerized = dataset1.copy()

for col_name in dataset1_numerized.columns:
    if dataset1_numerized[col_name].dtype == 'object':
        dataset1_numerized[col_name] = dataset1_numerized[col_name].astype('category')
        dataset1_numerized[col_name] = dataset1_numerized[col_name].cat.codes

dataset1_numerized
```

```python
correlation_matrix = dataset1_numerized.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot= True)
plt.title('Correlation matrix for Movie columns')
plt.show()
```
