```python
import pandas as pd
```

```python
#load dataset
callList = pd.read_excel(r"/Users/nemo/Downloads/Customer Call List.xlsx")
callList
```

```python
#drop any duplicates
callList = callList.drop_duplicates()
callList
```

```python
#dropping any unnecessary columns not useful for analysis
callList = callList.drop(columns = "Not_Useful_Column")
callList
```

```python
#strip unnecessay endings or values in the last name column

callList["Last_Name"] = callList["Last_Name"].str.strip("...")
callList["Last_Name"] = callList["Last_Name"].str.strip("/")
callList["Last_Name"] = callList["Last_Name"].str.rstrip("_")
callList
```

```python
#changing the format of all phone numbers 
callList["Phone_Number"] = callList["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex = True)
callList["Phone_Number"]
```

```python
#creating a lambda function to create a new format for the phone numbber column
callList["Phone_Number"] = callList["Phone_Number"].apply(lambda x: str(x))
callList["Phone_Number"] = callList["Phone_Number"].apply(lambda x: x[0:3] + "-" + x[3:6] + "-" + x[6:10])
callList["Phone_Number"]
```

```python
#removing nan values in the phone column
callList["Phone_Number"] = callList["Phone_Number"].str.replace("nan--", "")
callList["Phone_Number"] = callList["Phone_Number"].str.replace("Na--", "")
callList["Phone_Number"]
```

```python
callList
```

```python
#split the address coulmn into street adress, state and the zipcode
callList[["Street_address", "State", "Zipcode"]] = callList["Address"].str.split(",", n=2, expand = True)
callList
```

```python
#chnaging the format for yes and no for the paying_customer column
callList["Paying Customer"] = callList["Paying Customer"].str.replace("Yes", "Y")
callList["Paying Customer"] = callList["Paying Customer"].str.replace("No", "N")
callList["Paying Customer"] = callList["Paying Customer"].str.replace("N/a", " ")
callList["Paying Customer"] = callList["Paying Customer"].str.replace("Y", "Yes")
callList["Paying Customer"] = callList["Paying Customer"] = callList["Paying Customer"].str.replace("N", "No")
callList["Paying Customer"]
```

```python
#chnaging the format for yes and no for the do_not_contact column 
callList["Do_Not_Contact"] = callList["Do_Not_Contact"].str.replace("No", "N")
callList["Do_Not_Contact"] = callList["Do_Not_Contact"].str.replace("Yes", "Y")
callList["Do_Not_Contact"] = callList["Do_Not_Contact"].str.replace("NaN", "")
callList["Do_Not_Contact"] = callList["Do_Not_Contact"].str.replace("Y", "Yes")
callList["Do_Not_Contact"] = callList["Do_Not_Contact"].str.replace("N", "No")
callList["Do_Not_Contact"]
```

```python
#removing any null values in the table by suing replace and fillna
callList = callList.replace("N/a", "")
callList1 = callList.fillna("")
callList1
```

```python
#for loop function to drop any rows for customers who do not want to be contacted
for x in callList1.index:
    if callList1.loc[x, "Do_Not_Contact"] == "Yes":
      callList1.drop(x, inplace = True)
```

```python
#dropping any columns without phone numbers
for x in callList1.index:
    if callList1.loc[x, "Phone_Number"] == "":
      callList1.drop(x, inplace = True)
```

```python
#reset the index column
callList1 = callList1.reset_index(drop = True)
```
